IMAGE TO PENCIL SKETCH

```python
import cv2
from skimage import io
```

```python
#Importing the image
image = cv2.imread("Downloads/model.jpg")
img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
io.imshow(img)
```

```python
#Convert it into grayscale
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
io.imshow(gray_image)
```

```python
#Converting into Negative Image
inverted_image = 255 - gray_image
io.imshow(inverted_image)
```

```python
#Applying Gaussian Blur
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
io.imshow(blurred)
```

```python
inverted_blurred = 255 - blurred
io.imshow(inverted_blurred)
```

```python
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256)
io.imshow(pencil_sketch)
```
